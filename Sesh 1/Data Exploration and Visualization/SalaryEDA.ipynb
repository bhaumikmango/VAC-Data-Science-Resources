```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Salary_dataset.csv')
df.tail()
```

```python
df.drop('Unnamed: 0', axis='columns', inplace=True)
df
```

```python
import numpy as np
df['YearsExperience'] = np.round(df['YearsExperience']) #Rounding off to nearest integer
df['Salary'] = np.round(df['Salary'], decimals=2) #Rounding off to 2 decimal places 
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.columns
```

```python
plt.bar(df['YearsExperience'], df['Salary'], color = 'skyblue')
plt.plot(df['YearsExperience'], df['Salary'], color = 'red', marker = '*')
plt.title("Combined Bar and Line Chart")
plt.xlabel('Experience (in years)')
plt.ylabel('Salary (in USD)')
plt.show()
```
