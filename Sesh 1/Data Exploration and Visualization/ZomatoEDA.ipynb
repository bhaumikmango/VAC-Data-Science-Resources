```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head(10)
```

```python
df_country.columns
```

```python
df_country.isnull().sum()
```

```python
#merging these datasets with countrycode as primary key 
final=pd.merge(df, df_country, on='Country Code', how='left')
final.head(5)
```

```python
final.dtypes
```

```python
final.columns
```

```python
Country_names=final.Country.value_counts().index #.index gives the names of the attributes in order of their descending values
Country_val=final.Country.value_counts().values #.values gives the values of the attributes
final.Country.value_counts()
```

Here we conclude that Zomato is mainly active in India

```python
plt.pie(Country_val[:3], labels=Country_names[:3], autopct='%1.2f%%')
'''
only selecting top 3 countries using zomato, using the percent value 
till 2 digits after decimal point(if I don't use 2 in autopct parameter 
it will not provide the absolute value but the rounded offs)
'''
```

99.99% of all the services provided are in the top 3 countries of India, USA and UK

```python
ratings=final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings.head(5)
```

Observations:-
1. When rating is bw 4.5 - 4.9 --> excellent
2. When rating is bw 4.0 - 4.4 --> very good
3. When rating is bw 3.5 - 3.9 --> good
4. When rating is bw 2.5 - 3.4 --> average
5. When rating is bw 1.8 - 2.4 --> poor
6. rest all are unrated with a rating of 0

```python
import matplotlib
matplotlib.rcParams['figure.figsize']= (12,9)
sns.barplot(x='Aggregate rating', y='Rating count',hue='Rating color' ,data=ratings, palette=['gray', 'red', 'orange', 'yellow', 'green', 'green'])
```

1. Most are not rated
2. excluding unrated, most ratings lie between the range of 2.7 - 4.1

```python
sns.countplot(x='Rating color', data=ratings, palette=['gray', 'red', 'orange', 'yellow', 'green', 'green'])
#frequency of rating numbers not count
```

```python
#find countries that has given 0 rating
gh=final[final['Rating color']=='White']
gh.groupby(['Aggregate rating', 'Country']).size() 
```

Observation:- most of the 0 ratings are from India

```python
#create a pi chart for cities distribution
City_names=final.City.value_counts().index 
City_val=final.City.value_counts().values
plt.pie(City_val[:5], labels=City_names[:5], autopct='%1.2f%%')
```
